```python
# Importing the required libraries.
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

# Dataset :- Global Terrorism

```python
# Load the dataset.
terrorism = pd.read_csv(r"C:\Users\adity\OneDrive\Desktop\globalterrorismdb_0718dist.csv" ,encoding = "latin-1")
terrorism
```

```python
# Display basic information about the dataset
terrorism.info()
```

```python
terrorism.columns.values
```

```python
# Renaming  the columns name.
terrorism.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
terrorism.columns.values
```

# Droping Unnecessary Columns

```python
terrorism_df = terrorism[['Year','Month','Day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill','Wound','target1','Group','Targettype','Weapon']]
terrorism_df
```

```python
terrorism_df.info()
```

# Exploring Data

```python
# Checking for the null values.
terrorism_df.isnull().sum()
```

```python
 terrorism_df[['Wound','State','City', 'latitude','longitude','kill', 'target1']]= terrorism_df[['Wound','State','City', 'latitude','longitude','kill', 'target1']].fillna(0)
    
```

```python
terrorism_df.isnull().sum()
```

```python
# Checking for the duplictes.
terrorism_df.duplicated().sum()
```

```python
terrorism_df.drop_duplicates(inplace=True)
```

```python
terrorism_df.duplicated().sum()
```

```python
terrorism_df.describe()
```

```python
numeric_cols = terrorism_df[["Year","Month","Day","kill","Wound"]]
correlation = numeric_cols.corr()
correlation
```

# Visualisation

```python
sns.heatmap(correlation,xticklabels =correlation.columns, yticklabels = correlation.columns, annot = True, cmap = 'Oranges')
plt.show()
```

 From the above heat map we can day that it has a  strong positive correlation (0.44) between "kill" and "Wound", suggesting these two variables are highly related.

```python
# Time series analysis
# Plotting the trend of terrorist incidents over time

year = terrorism_df['Year'].unique()
years_count = terrorism_df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year, y = years_count,palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years From 1970 to 2017',fontsize=30)
plt.show()
```

Beginning in the early 2000s, there was a noticeable rise in the number of attacks, with <b>2014</b> seeing the largest peak.

```python
# Attack type analysis
plt.figure(figsize=(12, 10))
sns.countplot(x='Attacktype', data=terrorism_df, order=terrorism_df['Attacktype'].value_counts().index)
plt.title('Types of Terrorist Attacks')
plt.xlabel('Number of Incidents')
plt.ylabel('Attack Type')
plt.xticks(rotation=90)
plt.show()

```

Based on the bar graph showing the types of terrorist attacks, the most predominant type appears to be <b>Bombing/Explosion,</b> with an overwhelmingly larger number of incidents compared to other attack types.

```python
Most_attack_Country = terrorism_df.Country.value_counts()[:10]
Most_attack_Country
```

```python
# Plotting the graph
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

plt.figure(figsize=(10, 6))
Most_attack_Country.plot(kind='bar', color=colors)
plt.ylabel('Number of Attacks')
plt.xlabel('Country')
plt.title('Top 10 Countries by Number of Attacks')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Based on the bar graph showing the top 10 countries by number of attacks, <b>Iraq </b>has experienced the highest number of terrorist attacks compared to the other countries shown.

```python
Max_attack_group= terrorism_df.Group.value_counts()[1:10]
Max_attack_group
```

```python
# Extract group names and attack counts
group_names = Max_attack_group.index
attack_counts = Max_attack_group.values

# Plotting the graph
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

plt.figure(figsize=(10, 6))
plt.bar(group_names, attack_counts, color=colors)
plt.ylabel('Number of Attacks')
plt.xlabel('Terrorist Group')
plt.title('Top 10 Terrorist Groups by Number of Attacks')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

Based on the bar graph showing the top 10 terrorist groups by number of attacks, the <b>Taliban</b> has carried out the highest number of terrorist attacks compared to the other groups shown.

```python
# Target analysis
plt.figure(figsize=(10, 6))
sns.countplot(x='Targettype', data=terrorism_df, order=terrorism_df['Targettype'].value_counts().index)
plt.title('Types of Targets in Terrorist Incidents')
plt.xlabel('Number of Incidents')
plt.ylabel('Target Type')
plt.xticks(rotation=90)
plt.show()

```

<b>Private Citizens & Property </b>seem to be targets in terrorist incidents more often than other target categories, according to the graph depicting target types.

```python
weapon_counts = terrorism['Weapon'].value_counts()

# Select only the top 5 weapon types
top_5_weapon_counts = weapon_counts.head(5)

# Plotting the pie chart
plt.figure(figsize=(8,4))
plt.pie(top_5_weapon_counts, labels=top_5_weapon_counts.index, autopct='%1.1f%%')
plt.axis('equal') 
plt.title('Distribution of Top 5 Types of Weapons Used in Terrorist Attacks', fontsize = 20)

plt.show()
```

The most commonly used weapon in terrorist attacks is <b>Explosives</b> , accounting for <b>51.1%</b> of incidents.

```python
# Group the data by year and sum the number of people killed
deaths_per_year = terrorism_df.groupby('Year')['kill'].sum()

# Plotting the bar plot
plt.figure(figsize=(12, 6))
deaths_per_year.plot(kind='bar', color='cyan')
plt.title('Number of People killed Due to Attacks Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=90)
plt.show()
```

From the above graph we can conclude that most of the people were killed due to attack in <b>2014.</b>

```python
# Get the top 10 most affected cities
top_10_cities = terrorism_df['City'].value_counts().head(10)

# Print the top 10 most affected cities
print(top_10_cities)
```

```python
# Plotting the bar plot
plt.figure(figsize=(10, 6))
top_10_cities.plot(kind='bar', color='green')
plt.title('Top 10 Most Affected Cities by Number of Incidents')
plt.xlabel('City')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

From the above bar graph we can infer that the most affected cities was <b>Unknown</b> followed by <b>Baghdad.</b>

```python
# Group the data by attack type and sum the number of people killed
attack_kill_sum = terrorism_df.groupby('Attacktype')['kill'].sum().sort_values(ascending=False)

# Plotting the bar plot with improvements
plt.figure(figsize=(12, 8))
ax = attack_kill_sum.plot(kind='bar', color='#FF6347') 
ax.set_title("Number of People Killed by Attack Type", fontsize=20)
ax.set_xlabel("Attack Type", fontsize=15)
ax.set_ylabel("Number of People Killed", fontsize=15)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right', fontsize=12)  
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005), fontsize=12)  # Add data labels
plt.show()
```

The above bar graph shows that mostly people were killed by <b>Armed Assault</b> or <b>Bombing/Explosion.</b>

```python
terr_df=terrorism_df[['Group','Country','kill']]
df=terr_df.groupby(['Group','Country'],axis=0).sum().sort_values('kill',ascending=False).drop('Unknown').reset_index().head(10)
df
```

```python
# Group the data by year and region, then count the number of terrorist activities
pivot_table = terrorism_df.groupby(['Year', 'Region']).size().unstack(fill_value=0)

# Plotting the line plot
plt.figure(figsize=(20, 10))
sns.lineplot(data=pivot_table, dashes=False)
plt.title('Region Wise Terrorist Activities For Each Year', fontsize=25)
plt.ylabel('Number of Attacks', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Region', title_fontsize='15', fontsize='12')
plt.grid(True)
plt.show()
```
